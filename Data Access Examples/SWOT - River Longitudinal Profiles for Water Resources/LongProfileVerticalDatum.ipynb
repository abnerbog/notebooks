## 3 Define time period of interest

This will pull and show a timeseries of SWOT overpasses for a reach, by displaying a timeseries of water elevations.

```python
df=PullReachTimeseries(reachid)
```

```python
# Inspect downloaded data
df.head()
```

## 4 Convert WSE to NAVD88
We will use NOAA vdatum api (https://vdatum.noaa.gov/docs/services.html) to estimate the offset between SWOT EGM08 and NAVD88  

```python
# Extract representative x (longitude) and y (latitude) to use for vdatum api to convert vertical datum
x = df.p_lon.unique().item()
y = df.p_lat.unique().item()

# convert the first elevation in SWOT timeseries to NAVD88, meters. 
zin=df.iloc[0]['wse'] # first in series
swot_offset, zout = ChangeDatum(x, y, zin)  #zout=zin+swot_offset
# compute SWOT data in NAVD88 and convert to feet
df['wse_NAVD88'] = df['wse'] + swot_offset # apply offset
df['wse_NAVD88_ft'] = df['wse_NAVD88']/0.3048 # convert to feet
```

```python
# plot swot data as a timeseries (meters)
px.line(df,x='time_str',y='wse',
       labels={"time_str": "",
               "wse": "wse[m]"},
        markers=True)
```

```python
# plot swot data as a timeseries (feet)
px.line(df,x='time_str',y='wse_NAVD88_ft',
       labels={"time_str": "",
               "wse_NAVD88_ft": "wse_NAVD88 [ft]"},
        markers=True)
```

## Pull longitudinal profile

```python
longdf=PullLongitudinalProfile(reachid,tlong)
```

```python
# Convert to NAVD88
longdf["wse"] = longdf["wse"] + swot_offset
```

```python
px.line(longdf,x='p_dist_out',y='wse',
       labels={"p_dist_out": "Distance to outlet [m]",
               "wse": "wse[m]"},
        markers=True)
```

The distance to outlet data are not easy to parse

Let's plot distance as kilometers to Tanana confluence instead 

```python
longdf['dist_up_conf']=(longdf['p_dist_out']-longdf['p_dist_out'].min())/1000.
```

```python
px.line(longdf,x='dist_up_conf',y='wse',
       labels={"dist_up_conf": "Distance to Tanana confluence [km]",
               "wse": "wse[m]"},
        markers=True)
```

```python
# convert to English units
longdf['wse [ft]'] = longdf['wse']/0.3048
longdf['dist_up_conf [mi]'] = longdf['dist_up_conf']*0.62
```

```python
px.line(longdf,x='dist_up_conf [mi]',y='wse [ft]',
       labels={"dist_up_conf [mi]": "Distance to Tanana confluence [mi]",
               "wse": "wse[feet]"},
        markers=True)
```
