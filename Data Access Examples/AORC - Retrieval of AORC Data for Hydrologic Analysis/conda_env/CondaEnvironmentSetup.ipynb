# Setting Up A Custom Conda Environment for Running AORC Jupyter Notebooks

The AORC notebooks were developed in the *CIROH 2i2c JupyterHub* environment using the default *New Pangeo base image* and are fully compatible with its pre-installed Python modules. To ensure the smoothest experience when running the notebooks on *CIROH 2i2c JupyterHub* or other available JupyterHub environments, including *CUAHSI JupyterHub* and *CyberGIS-Jupyter for Water*, you need to first **create a custom conda environment** in your home directory (see below) to ensure all required Python modules are available.   

💡**Note:** **You only need to create this conda environment once per JupyterHub environment—the first time you run the notebooks**.

### 🛠️ To **create** a custom conda environment, run the following cell.
Creating a new conda environment may take some time and users need to be patient. Please wait until you see `✅ Conda environment 'aorc'  successfully created ...` at the end.

```python
%%bash
chmod +x setup_aorc_jupyterhub_conda_env.sh
./setup_aorc_jupyterhub_conda_env.sh
```

You can check the conda environment list by running the following cell to verify the environment was created:

```python
%%bash
conda env list
```

- To use this conda environment for running a notebook, switch the kernel to **Python [cond env:.conda-aorc]** if this kernel is in the list. Otherwise, use **Python3 (AORC)**. The *Switch kernel* button is located at the top right corner of each notebook.
- If the aorc kernel does not appear as an option immediately, you may need to **restart the kernel** to refresh the list of available kernels. To do this, go to the *Kernel* tab and select *Restart Kernel...*
- If you encounter package conflict errors, make sure you have selected the *aorc* kernel when running AORC notebooks on JupyterHub environments. 

### ❌ (Optional) If you want to **delete** the *aorc* conda environment, run the following cell.
It may take some time and users need to be patient. Please wait until you see `✅ Environment and kernel cleanup complete` at the end.

```python
%%bash
chmod +x delete_aorc_jupyterhub_conda_env.sh
./delete_aorc_jupyterhub_conda_env.sh
```
