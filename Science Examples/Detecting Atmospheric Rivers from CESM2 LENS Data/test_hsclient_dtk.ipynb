# Grab a notebook from a HydroShare resource: a test of knowledge  
DTK using the hsclient to grab the newer version of the atmospheric river notebook from the resource:
https://hydroshare.org/resource/6920ead7e39c4c7faba951c1417fb4b6/


```python
from hsclient import HydroShare

hs = HydroShare()
hs.sign_in()
```

```python
res_identifier = "6920ead7e39c4c7faba951c1417fb4b6"
# Get an existing resource using its identifier
existing_resource = hs.resource(res_identifier)
print(f'Just retrieved the resource with ID: {res_identifier}')
```

```python
# Get the list of all aggregations in the resource
aggregations = existing_resource.aggregations()
```
